```python
import torch
from ultralytics import YOLO
```

```python
print(f"GPU: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Device: {torch.cuda.get_device_name(0)}")
```

```python
model = YOLO('yolo11n-pose.pt')
```

```python
results = model.train(
    data='data.yaml',
    epochs=200,
    imgsz=1024,      
    batch=4,         
    rect=True,       
    mosaic=0.0,      
    close_mosaic=0,
    degrees=0.0,     # no rotation
    flipud=0.0,      # no reverse
    fliplr=0.5,      # left-right mirror
    device=0
)
```
