```python
import pandas as pd
```

```python
new_dicom_df = pd.read_csv("new-dicom.csv")
old_dicom_df = pd.read_csv("old-dicom.csv")
```

```python
result = pd.concat([new_dicom_df, old_dicom_df], axis=0, ignore_index=True)
result.drop_duplicates()
result
```

```python
def find_duplicate_rows(df1, df2):
    # İki DataFrame arasındaki ortak satırları bul (Inner Join)
    # how='inner': Sadece her iki df'de de olan satırları getirir
    common_rows = pd.merge(df1, df2, how='inner')

    # Kaç tane aynı satır olduğunu hesapla
    count = len(common_rows)

    if count > 0:
        print(f"Total duplicate rows found: {count}")
        print("\nDuplicate Rows:")
        print(common_rows)
    else:
        print("No duplicate rows found between the two DataFrames.")
    
    return common_rows, count
```

```python
find_duplicate_rows(new_dicom_df, old_dicom_df)
```
