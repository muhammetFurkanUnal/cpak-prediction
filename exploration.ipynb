```python
import pandas as pd
```

```python
images = []
with open("out.txt", "r") as file:
	for line in file:
		images.append(line.strip())
```

```python
csv_path = "./deeplabcut/cpak-furkan-2026-02-17/labeled-data/images/CollectedData_murkanthedestroyer.csv"
df = pd.read_csv(csv_path, header=[0,1,2], index_col=0)
```

```python
df.head()
```

```python
# find only labeled images from lera dataset
path_prefix = "labeled-data/images/"

only_labeled_lera = df[df.index.isin([path_prefix+image for image in images])]
print(len(only_labeled_lera))
only_labeled_lera.head(20)

```

```python
with open("only_lera.csv", "w") as file:
	file.write(only_labeled_lera.to_csv())
```
